```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras import layers, Model
from sklearn.model_selection import train_test_split
from keras import Model
from keras.layers import Conv2D, PReLU, BatchNormalization, Flatten
from keras.layers import UpSampling2D, LeakyReLU, Dense, Input, add
from tqdm import tqdm
from keras.applications import VGG19
import tensorflow as tf

#########################################################################

# Set memory growth for GPUs to avoid memory allocation issues
gpus = tf.config.experimental.list_physical_devices('GPU')
if gpus:
    try:
        # Set memory growth for each GPU to allow dynamic memory allocation
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
        print(f"Using {len(gpus)} GPU(s).")
    except RuntimeError as e:
        print(e)
else:
    print("No GPUs detected. Using CPU.")

# If you want to use the first GPU explicitly:
# tf.config.set_visible_devices(gpus[0], 'GPU')

def res_block(ip):
    res_model = Conv2D(64, (3, 3), padding="same")(ip)
    res_model = BatchNormalization(momentum=0.5)(res_model)
    res_model = PReLU(shared_axes=[1, 2])(res_model)

    res_model = Conv2D(64, (3, 3), padding="same")(res_model)
    res_model = BatchNormalization(momentum=0.5)(res_model)

    return add([ip, res_model])

def upscale_block(ip):
    up_model = Conv2D(256, (3, 3), padding="same")(ip)
    up_model = UpSampling2D(size=2)(up_model)
    up_model = PReLU(shared_axes=[1, 2])(up_model)

    return up_model

# Generator model
def create_gen(gen_ip, num_res_block):
    layers = Conv2D(64, (9, 9), padding="same")(gen_ip)
    layers = PReLU(shared_axes=[1, 2])(layers)

    temp = layers

    for i in range(num_res_block):
        layers = res_block(layers)

    layers = Conv2D(64, (3, 3), padding="same")(layers)
    layers = BatchNormalization(momentum=0.5)(layers)
    layers = add([layers, temp])

    layers = upscale_block(layers)
    layers = upscale_block(layers)

    op = Conv2D(3, (9, 9), padding="same")(layers)

    return Model(inputs=gen_ip, outputs=op)

# Discriminator block that will be used to construct the discriminator
def discriminator_block(ip, filters, strides=1, bn=True):
    disc_model = Conv2D(filters, (3, 3), strides=strides, padding="same")(ip)
    if bn:
        disc_model = BatchNormalization(momentum=0.8)(disc_model)

    disc_model = LeakyReLU(alpha=0.2)(disc_model)

    return disc_model

# Discriminator, as described in the original paper
def create_disc(disc_ip):
    df = 64

    d1 = discriminator_block(disc_ip, df, bn=False)
    d2 = discriminator_block(d1, df, strides=2)
    d3 = discriminator_block(d2, df * 2)
    d4 = discriminator_block(d3, df * 2, strides=2)
    d5 = discriminator_block(d4, df * 4)
    d6 = discriminator_block(d5, df * 4, strides=2)
    d7 = discriminator_block(d6, df * 8)
    d8 = discriminator_block(d7, df * 8, strides=2)

    d8_5 = Flatten()(d8)
    d9 = Dense(df * 16)(d8_5)
    d10 = LeakyReLU(alpha=0.2)(d9)
    validity = Dense(1, activation='sigmoid')(d10)

    return Model(disc_ip, validity)

# VGG19
def build_vgg(hr_shape):
    vgg = VGG19(weights="imagenet", include_top=False, input_shape=hr_shape)
    return Model(inputs=vgg.inputs, outputs=vgg.layers[10].output)

# Combined model
def create_comb(gen_model, disc_model, vgg, lr_ip, hr_ip):
    gen_img = gen_model(lr_ip)
    gen_features = vgg(gen_img)
    disc_model.trainable = False
    validity = disc_model(gen_img)
    return Model(inputs=[lr_ip, hr_ip], outputs=[validity, gen_features])

# Load first n number of images (to train on a subset of all images)
n = 500
lr_list = os.listdir("./lr_images")[:n]

lr_images = []
for img in lr_list:
    img_lr = cv2.imread("./lr_images/" + img)
    if img_lr is not None:
        img_lr = cv2.cvtColor(img_lr, cv2.COLOR_BGR2RGB)
        lr_images.append(img_lr)
    else:
        print(f"Failed to load image: {img}")

hr_list = os.listdir("./hr_images")[:n]

hr_images = []
for img in hr_list:
    img_hr = cv2.imread("./hr_images/" + img)
    if img_hr is not None:
        img_hr = cv2.cvtColor(img_hr, cv2.COLOR_BGR2RGB)
        hr_images.append(img_hr)
    else:
        print(f"Failed to load image: {img}")

lr_images = np.array(lr_images)
hr_images = np.array(hr_images)

# Sanity check, view a few images
import random
image_number = random.randint(0, len(lr_images) - 1)
plt.figure(figsize=(12, 6))
plt.subplot(121)
plt.imshow(np.reshape(lr_images[image_number], (32, 32, 3)))
plt.subplot(122)
plt.imshow(np.reshape(hr_images[image_number], (128, 128, 3)))
plt.show()

# Scale values
lr_images = lr_images / 255.
hr_images = hr_images / 255.

# Split to train and test
lr_train, lr_test, hr_train, hr_test = train_test_split(lr_images, hr_images, test_size=0.33, random_state=42)

hr_shape = (hr_train.shape[1], hr_train.shape[2], hr_train.shape[3])
lr_shape = (lr_train.shape[1], lr_train.shape[2], lr_train.shape[3])

lr_ip = Input(shape=lr_shape)
hr_ip = Input(shape=hr_shape)
generator=tf.keras.models.load_model('./g_model.h5')
discriminator=tf.keras.models.load_model('./d_model.h5')
from keras.models import load_model
from keras.losses import MeanSquaredError as mse
gan_model = load_model('./gan_model.h5', custom_objects={'mse': mse})

# Use GPU for model creation
with tf.device('/GPU:0'):
    generator = create_gen(lr_ip, num_res_block=16)
    generator.summary()

    discriminator = create_disc(hr_ip)
    discriminator.compile(loss="binary_crossentropy", optimizer="adam", metrics=['accuracy'])
    discriminator.summary()

    # Build VGG model
    vgg = build_vgg((128, 128, 3))
    vgg.trainable = False

    # Create GAN model
    gan_model = create_comb(generator, discriminator, vgg, lr_ip, hr_ip)

    # Compile GAN model with adversarial loss and content loss
    gan_model.compile(loss=["binary_crossentropy", "mse"], loss_weights=[1e-3, 1], optimizer="adam")
    gan_model.summary()

```

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras import layers, Model
from sklearn.model_selection import train_test_split
from keras import Model
from keras.layers import Conv2D, PReLU, BatchNormalization, Flatten
from keras.layers import UpSampling2D, LeakyReLU, Dense, Input, add
from tqdm import tqdm
from keras.applications import VGG19
import tensorflow as tf
n = 500
lr_list = os.listdir("./lr_images")[:n]

lr_images = []
for img in lr_list:
    img_lr = cv2.imread("./lr_images/" + img)
    if img_lr is not None:
        img_lr = cv2.cvtColor(img_lr, cv2.COLOR_BGR2RGB)
        lr_images.append(img_lr)
    else:
        print(f"Failed to load image: {img}")

hr_list = os.listdir("./hr_images")[:n]

hr_images = []
for img in hr_list:
    img_hr = cv2.imread("./hr_images/" + img)
    if img_hr is not None:
        img_hr = cv2.cvtColor(img_hr, cv2.COLOR_BGR2RGB)
        hr_images.append(img_hr)
    else:
        print(f"Failed to load image: {img}")

lr_images = np.array(lr_images)
hr_images = np.array(hr_images)

# Sanity check, view a few images
import random
image_number = random.randint(0, len(lr_images) - 1)
plt.figure(figsize=(12, 6))
plt.subplot(121)
plt.imshow(np.reshape(lr_images[image_number], (32, 32, 3)))
plt.subplot(122)
plt.imshow(np.reshape(hr_images[image_number], (128, 128, 3)))
plt.show()

# Scale values
lr_images = lr_images / 255.
hr_images = hr_images / 255.

# Split to train and test
lr_train, lr_test, hr_train, hr_test = train_test_split(lr_images, hr_images, test_size=0.33, random_state=42)

hr_shape = (hr_train.shape[1], hr_train.shape[2], hr_train.shape[3])
lr_shape = (lr_train.shape[1], lr_train.shape[2], lr_train.shape[3])

lr_ip = Input(shape=lr_shape)
hr_ip = Input(shape=hr_shape)
generator=tf.keras.models.load_model('./g_model.h5')
discriminator=tf.keras.models.load_model('./d_model.h5')
from keras.models import load_model
from keras.losses import MeanSquaredError as mse
gan_model = load_model('./gan_model.h5', custom_objects={'mse': mse})
def build_vgg(hr_shape):
    vgg = VGG19(weights="imagenet", include_top=False, input_shape=hr_shape)
    return Model(inputs=vgg.inputs, outputs=vgg.layers[10].output)

# Use GPU for model creation
with tf.device('/GPU:0'):
    generator.summary()

    discriminator.compile(loss="binary_crossentropy", optimizer="adam", metrics=['accuracy'])
    discriminator.summary()

    vgg = build_vgg((128, 128, 3))
    vgg.trainable = False

    # Compile GAN model with adversarial loss and content loss
    gan_model.compile(loss=["binary_crossentropy", "mse"], loss_weights=[1e-3, 1], optimizer="adam")
    gan_model.summary()

```

```python
import tensorflow.keras.backend as K
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm

batch_size = 16
train_lr_batches = []
train_hr_batches = []

# Prepare training batches
for it in range(int(hr_train.shape[0] / batch_size)):
    start_idx = it * batch_size
    end_idx = start_idx + batch_size
    train_hr_batches.append(hr_train[start_idx:end_idx])
    train_lr_batches.append(lr_train[start_idx:end_idx])

epochs = 50
# Enumerate training over epochs
for e in range(epochs):
    fake_label = np.zeros((batch_size, 1))  # Assign a label of 0 to all fake (generated images)
    real_label = np.ones((batch_size, 1))  # Assign a label of 1 to all real images.

    # Create empty lists to store gen and disc losses
    g_losses = []
    d_losses = []

    # Enumerate training over batches
    for b in tqdm(range(len(train_hr_batches))):
        lr_imgs = train_lr_batches[b]  # Fetch a batch of LR images for training
        hr_imgs = train_hr_batches[b]  # Fetch a batch of HR images for training

        fake_imgs = generator.predict_on_batch(lr_imgs)  # Generate fake images

        # First, train the discriminator on fake and real HR images
        discriminator.trainable = True
        d_loss_gen = discriminator.train_on_batch(fake_imgs, fake_label)
        d_loss_real = discriminator.train_on_batch(hr_imgs, real_label)

        # Now, train the generator by fixing discriminator as non-trainable
        discriminator.trainable = False

        # Average the discriminator loss for reporting purposes
        d_loss = 0.5 * np.add(d_loss_gen, d_loss_real)

        # Extract VGG features to calculate the content loss
        image_features = vgg.predict(hr_imgs)

        # Train the generator via GAN
        # We have two losses: adversarial loss and content (VGG) loss
        outputs = gan_model.train_on_batch([lr_imgs, hr_imgs], [real_label, image_features])

        # Unpack the loss values
        g_loss = outputs[0]  # Adversarial loss
        content_loss = outputs[1]  # Content loss (VGG features)

        # Save losses to the list
        d_losses.append(d_loss)
        g_losses.append(g_loss)

    # Calculate the average losses for generator and discriminator
    g_losses = np.array(g_losses)
    d_losses = np.array(d_losses)

    g_loss = np.sum(g_losses, axis=0) / len(g_losses)
    d_loss = np.sum(d_losses, axis=0) / len(d_losses)

    # Report the progress during training
    print(f"Epoch: {e + 1}, g_loss: {g_loss}, d_loss: {d_loss}")

    # Check the generator output (for example, by generating a sample image)
    if (e + 1) % 1 == 0:  # Adjust frequency as needed (e.g., after every epoch)
        sample_lr = train_lr_batches[5]  # Get a sample batch of LR images
        generated_image = generator.predict_on_batch(sample_lr)  # Generate HR image
        original_hr = train_hr_batches[5]  # Get the corresponding HR image
    
        # Display the generated image and original HR image side by side
        plt.figure(figsize=(12, 6))  # Adjust the figure size as needed
        # Plot generated image
        plt.subplot(1, 2, 1)
        plt.imshow(generated_image[5])  # Display the first image in the batch
        plt.axis('off')
        plt.title(f"Generated Image at Epoch {e + 1}")
        
        # Plot original HR image
        plt.subplot(1, 2, 2)
        plt.imshow(original_hr[5])  # Display the first image in the batch of original HR
        plt.axis('off')
        plt.title("Original HR Image")
        
        plt.show()

    # Save the models periodically (every 2 epochs as in your original code)
    if (e + 1) % 1 == 0:
        generator.save("g_model.h5")
        discriminator.save("d_model.h5")
        gan_model.save("gan_model.h5")

    # Clear memory at the end of each epoch
    K.clear_session()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
import cv2
from tqdm import tqdm
from tensorflow.keras.models import load_model

# Load the generator model (assuming you have the saved model file)
generator = generator  # Example model file

# Assuming you have 'lr_test' and 'hr_test' (low-res and high-res test data) as numpy arrays
# Prepare testing data by dividing them into batches of size 'batch_size'
batch_size = 1
test_lr_batches = []
test_hr_batches = []

# Divide data into batches
for it in range(int(hr_test.shape[0] / batch_size)):
    start_idx = it * batch_size
    end_idx = start_idx + batch_size
    test_lr_batches.append(lr_test[start_idx:end_idx])
    test_hr_batches.append(hr_test[start_idx:end_idx])

# Check the number of test batches
print("Number of test batches:", len(test_hr_batches))

# Generate fake images for each batch in the test set
fake_images = []
for b in tqdm(range(len(test_lr_batches))):
    lr_imgs = test_lr_batches[b]
    fake_imgs_batch = generator.predict_on_batch(lr_imgs)
    fake_images.append(fake_imgs_batch)

# Evaluate the performance using MSE and PSNR
mse = []
psnr = []

for i in range(len(test_hr_batches)):
    real_hr = test_hr_batches[i]
    fake_hr = fake_images[i]

    # Flatten the images before calculating MSE
    real_hr_flat = real_hr.flatten()
    fake_hr_flat = fake_hr.flatten()

    # Calculate Mean Squared Error (MSE)
    mse_batch = mean_squared_error(real_hr_flat, fake_hr_flat)
    mse.append(mse_batch)

    # Calculate Peak Signal-to-Noise Ratio (PSNR)
    mse_value = mean_squared_error(real_hr_flat, fake_hr_flat)
    psnr_batch = 10 * np.log10((np.max(real_hr) ** 2) / mse_value)
    psnr.append(psnr_batch)

# Average MSE and PSNR
avg_mse = np.mean(mse)
avg_psnr = np.mean(psnr)

print("Average MSE:", avg_mse)
print("Average PSNR:", avg_psnr)

# Visualize some of the generated images and compare them to the ground truth
num_images_to_display = 5  # Number of images to display
num_batches = len(test_hr_batches)

# Ensure we don't go out of bounds
num_images_to_display = min(num_images_to_display, num_batches)

for i in range(num_images_to_display):
    plt.figure(figsize=(12, 6))

    # Show original image (Ground Truth)
    plt.subplot(1, 2, 1)
    # Assuming the images are in the range [0, 1], multiply by 255 to scale to [0, 255]
    plt.imshow(test_hr_batches[i][0] )  # Assuming first image in the batch
    plt.title("Ground Truth")
    plt.axis('off')

    # Show generated image (Fake Image)
    plt.subplot(1, 2, 2)
    # Assuming the images are in the range [0, 1], multiply by 255 to scale to [0, 255]
    plt.imshow(fake_images[i][0])  # First image in the batch
    plt.title("Generated Image")
    plt.axis('off')

    plt.show()

# Save generated images for further inspection
for i in range(len(fake_images)):
    fake_image = fake_images[i][0]  # First image in the batch
    fake_image_path = f"generated_image_{i+1}.png"
    # Assuming images are normalized to [0, 1], multiply by 255 before saving
    cv2.imwrite(fake_image_path, fake_image * 255)  # Save in [0, 255] range
    print(f"Saved: {fake_image_path}")
```
